1. **The Probability Space**
A probability space is defined by the triplet $(\Omega, \mathscr{F}, P)$.

**Sample Space ($\Omega$)**: The set of all possible outcomes.
$$\Omega = \{\omega_1, \omega_2, \omega_3, \omega_4\}$$
**Sigma-Algebra ($\mathscr{F}$)**: The power set $\mathscr{P}(\Omega)$, which includes all $2^4 = 16$ subsets of $\Omega$:
$$\mathscr{F} = \{\emptyset, \{\omega_1\}, \{\omega_2\}, \dots, \{\omega_1, \omega_2\}, \dots, \Omega\}$$
**Probability Measure ($P$)**: Defined by the sum of the probabilities of the outcomes in any set $A \in \mathscr{F}$. Given the initial values, for any $A \in \mathscr{F}$:
$$P(A) = \sum_{\omega_i \in A} P(\{\omega_i\})$$

2. **The $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$**
The $\sigma$-algebra generated by $Y$, denoted as $\sigma(Y)$, consists of the preimages of all subsets of the range of $Y$.
Since $Y(\omega_1) = 0, Y(\omega_2) = 0, Y(\omega_3) = 1, Y(\omega_4) = 1$:
$Y=0$ corresponds to the set $\{\omega_1, \omega_2\}$
$Y=1$ corresponds to the set $\{\omega_3, \omega_4\}$
The $\sigma$-algebra must contain these sets, the empty set, and the whole sample space.
$$\sigma(Y) = \{\emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}, \Omega\}$$

3. **Why $E[X \mid Y]$ must be $\sigma(Y)$-measurable**
By definition, the conditional expectation $E[X \mid Y]$ is a random variable that is measurable with respect to the $\sigma$-algebra generated by the conditioning variable $Y$.
Because $\sigma(Y)$ only contains the sets $\emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}$, and $\Omega$, any $\sigma(Y)$-measurable function cannot distinguish between individual outcomes within those atomic sets. It cannot differentiate $\omega_1$ from $\omega_2$, nor $\omega_3$ from $\omega_4$. Therefore, the function $E[X \mid Y]$ must yield a constant value on the set $\{\omega_1, \omega_2\}$ and a (potentially different) constant value on the set $\{\omega_3, \omega_4\}$.

4. **Compute $E[X \mid Y]$ explicitly**
We calculate the expected value of $X$ given each possible value of $Y$.
First, find the probabilities of the conditions:
$P(Y=0) = P(\{\omega_1, \omega_2\}) = 0.50 + 0.10 = 0.60$
$P(Y=1) = P(\{\omega_3, \omega_4\}) = 0.20 + 0.20 = 0.40$
Now compute the conditional expectations:
$$E[X \mid Y=0] = \frac{X(\omega_1)P(\omega_1) + X(\omega_2)P(\omega_2)}{P(Y=0)} = \frac{0(0.50) + 1(0.10)}{0.60} = \frac{0.10}{0.60} = \frac{1}{6}$$
$$E[X \mid Y=1] = \frac{X(\omega_3)P(\omega_3) + X(\omega_4)P(\omega_4)}{P(Y=1)} = \frac{0(0.20) + 1(0.20)}{0.40} = \frac{0.20}{0.40} = \frac{1}{2}$$
Expressing this as a single random variable using indicator functions matches the desired result:
$$E[X \mid Y] = \frac{1}{6}\mathbb{1}_{\{Y=0\}} + \frac{1}{2}\mathbb{1}_{\{Y=1\}}$$

5. **Verify the defining property of conditional expectation**
We must check that $\int_A E[X \mid Y] dP = \int_A X dP$ for all $A \in \sigma(Y)$. We only need to check the non-trivial sets $A_1 = \{\omega_1, \omega_2\}$ and $A_2 = \{\omega_3, \omega_4\}$.
For $A_1 = \{\omega_1, \omega_2\}$ (where $Y=0$):
LHS: $\int_{A_1} E[X \mid Y] dP = \frac{1}{6} \cdot P(A_1) = \frac{1}{6}(0.60) = 0.10$
RHS: $\int_{A_1} X dP = X(\omega_1)P(\omega_1) + X(\omega_2)P(\omega_2) = 0(0.50) + 1(0.10) = 0.10$
LHS = RHS.
For $A_2 = \{\omega_3, \omega_4\}$ (where $Y=1$):
LHS: $\int_{A_2} E[X \mid Y] dP = \frac{1}{2} \cdot P(A_2) = \frac{1}{2}(0.40) = 0.20$
RHS: $\int_{A_2} X dP = X(\omega_3)P(\omega_3) + X(\omega_4)P(\omega_4) = 0(0.20) + 1(0.20) = 0.20$
LHS = RHS.

6. **Find the Conditional PMF $p_{X \mid Y}(x \mid y)$**
We calculate $P(X=x \mid Y=y) = \frac{P(X=x, Y=y)}{P(Y=y)}$.
$p_{X \mid Y}(0 \mid 0) = \frac{P(\{\omega_1\})}{P(Y=0)} = \frac{0.50}{0.60} = \frac{5}{6}$
$p_{X \mid Y}(1 \mid 0) = \frac{P(\{\omega_2\})}{P(Y=0)} = \frac{0.10}{0.60} = \frac{1}{6}$
$p_{X \mid Y}(0 \mid 1) = \frac{P(\{\omega_3\})}{P(Y=1)} = \frac{0.20}{0.40} = \frac{1}{2}$
$p_{X \mid Y}(1 \mid 1) = \frac{P(\{\omega_4\})}{P(Y=1)} = \frac{0.20}{0.40} = \frac{1}{2}$

8. **Calculate the Mean Square Error (MSE)**
The MSE of predicting $X$ using its conditional expectation is $E[(X - E[X \mid Y])^2]$. This is equivalent to $E[\text{Var}(X \mid Y)]$.
$\text{Var}(X \mid Y=0) = p(1-p) = \left(\frac{1}{6}\right)\left(\frac{5}{6}\right) = \frac{5}{36}$
$\text{Var}(X \mid Y=1) = p(1-p) = \left(\frac{1}{2}\right)\left(\frac{1}{2}\right) = \frac{1}{4}$
$$E[\text{Var}(X \mid Y)] = \frac{5}{36} P(Y=0) + \frac{1}{4} P(Y=1)$$
$$E[\text{Var}(X \mid Y)] = \frac{5}{36}(0.60) + \frac{1}{4}(0.40) = \left(\frac{5}{36} \cdot \frac{3}{5}\right) + \left(\frac{1}{4} \cdot \frac{2}{5}\right) = \frac{1}{12} + \frac{1}{10} = \frac{11}{60} \approx 0.1833$$

9. **Change in MSE and the Law of Total Variance**
If the bank ignores $Y$ and uses the global average $E[X] = 0.30$, the new MSE is simply the variance of $X$. Since $X$ is a Bernoulli variable with $p = 0.30$:
$$\text{Var}(X) = 0.30(1 - 0.30) = 0.21 = \frac{21}{100}$$
Relationship to Law of Total Variance:
The Law of Total Variance states: $\text{Var}(X) = E[\text{Var}(X \mid Y)] + \text{Var}(E[X \mid Y])$.
The MSE using the risk class $Y$ is $E[\text{Var}(X \mid Y)] = \frac{11}{60} \approx 0.1833$.
The MSE ignoring $Y$ is the total variance $\text{Var}(X) = 0.21$.
By ignoring $Y$, the MSE increases by exactly $\text{Var}(E[X \mid Y])$, which represents the variance "explained" by the customer's risk class.

10. **Entropy and Mutual Information**
a) Entropy $H(X)$:
First, the marginal probabilities: $P(X=0) = 0.70$ and $P(X=1) = 0.30$.
$$H(X) = - \sum p(x) \log_2 p(x)$$
$$H(X) = - [0.7 \log_2(0.7) + 0.3 \log_2(0.3)] \approx - [0.7(-0.5146) + 0.3(-1.737)] \approx 0.8813 \text{ bits}$$
b) Conditional Entropy $H(X \mid Y)$:
$$H(X \mid Y=0) = - \left[ \frac{5}{6}\log_2\left(\frac{5}{6}\right) + \frac{1}{6}\log_2\left(\frac{1}{6}\right) \right] \approx 0.6500 \text{ bits}$$
$$H(X \mid Y=1) = - \left[ \frac{1}{2}\log_2\left(\frac{1}{2}\right) + \frac{1}{2}\log_2\left(\frac{1}{2}\right) \right] = 1.0 \text{ bits}$$
$$H(X \mid Y) = P(Y=0)H(X \mid Y=0) + P(Y=1)H(X \mid Y=1)$$
$$H(X \mid Y) = 0.6(0.6500) + 0.4(1.0) = 0.390 + 0.40 = 0.790 \text{ bits}$$
c) Mutual Information $I(X; Y)$:
$$I(X; Y) = H(X) - H(X \mid Y)$$
$$I(X; Y) = 0.8813 - 0.790 = 0.0913 \text{ bits}$$
d) Percentage of uncertainty removed:
$$\text{Percentage} = \frac{I(X; Y)}{H(X)} \times 100$$
$$\text{Percentage} = \frac{0.0913}{0.8813} \times 100 \approx 10.36\%$$
Knowing the customer's risk class removes approximately 10.36% of the uncertainty regarding whether a transaction is fraudulent.

1. **Construct an independent random vector**
We want to define a new random vector $\mathbf{U}$ as a linear combination of $\mathbf{X}$ and $\mathbf{Y}$ such that $\mathbf{U}$ and $\mathbf{Y}$ are uncorrelated (and therefore independent, since they are jointly Gaussian). Let:
$$\mathbf{U} = \mathbf{X} - \mathbf{A}\mathbf{Y}$$
where $\mathbf{A}$ is a constant matrix we need to determine.
We set the covariance between $\mathbf{U}$ and $\mathbf{Y}$ to zero:
$$\text{Cov}(\mathbf{U}, \mathbf{Y}) = \text{Cov}(\mathbf{X} - \mathbf{A}\mathbf{Y}, \mathbf{Y}) = \mathbf{0}$$
Using the properties of covariance ($\text{Cov}(\mathbf{M}, \mathbf{N}) = E[(\mathbf{M} - E[\mathbf{M}])(\mathbf{N} - E[\mathbf{N}])^T]$):
$$\text{Cov}(\mathbf{X}, \mathbf{Y}) - \mathbf{A}\text{Cov}(\mathbf{Y}, \mathbf{Y}) = \mathbf{0}$$
$$\boldsymbol{\Sigma_{XY}} - \mathbf{A}\boldsymbol{\Sigma_{YY}} = \mathbf{0}$$
Solving for $\mathbf{A}$, we get:
$$\mathbf{A} = \boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1}$$
So, our independent random vector is $\mathbf{U} = \mathbf{X} - \boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1}\mathbf{Y}$. Because $\text{Cov}(\mathbf{U}, \mathbf{Y}) = \mathbf{0}$, $\mathbf{U}$ and $\mathbf{Y}$ are independent.

2. **Find the mean and variance of $\mathbf{U}$**
Now, we calculate the expected value (mean) and variance (covariance matrix) of $\mathbf{U}$.
Mean of $\mathbf{U}$:
$$E[\mathbf{U}] = E[\mathbf{X}] - \mathbf{A}E[\mathbf{Y}]$$
$$E[\mathbf{U}] = \boldsymbol{\mu_X} - \boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1}\boldsymbol{\mu_Y}$$
Variance of $\mathbf{U}$:
$$\text{Var}(\mathbf{U}) = \text{Cov}(\mathbf{U}, \mathbf{U}) = \text{Cov}(\mathbf{X} - \mathbf{A}\mathbf{Y}, \mathbf{X} - \mathbf{A}\mathbf{Y})$$
$$\text{Var}(\mathbf{U}) = \text{Var}(\mathbf{X}) - \text{Cov}(\mathbf{X}, \mathbf{Y})\mathbf{A}^T - \mathbf{A}\text{Cov}(\mathbf{Y}, \mathbf{X}) + \mathbf{A}\text{Var}(\mathbf{Y})\mathbf{A}^T$$
Substitute the known covariance matrices and $\mathbf{A} = \boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1}$:
$$\text{Var}(\mathbf{U}) = \boldsymbol{\Sigma_{XX}} - \boldsymbol{\Sigma_{XY}}(\boldsymbol{\Sigma_{YY}}^{-1})^T\boldsymbol{\Sigma_{YX}} - (\boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1})\boldsymbol{\Sigma_{YX}} + (\boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1})\boldsymbol{\Sigma_{YY}}(\boldsymbol{\Sigma_{YY}}^{-1})^T\boldsymbol{\Sigma_{YX}}$$
Since covariance matrices are symmetric, $\boldsymbol{\Sigma_{YY}}^T = \boldsymbol{\Sigma_{YY}}$ and $(\boldsymbol{\Sigma_{YY}}^{-1})^T = \boldsymbol{\Sigma_{YY}}^{-1}$. This simplifies to:
$$\text{Var}(\mathbf{U}) = \boldsymbol{\Sigma_{XX}} - \boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1}\boldsymbol{\Sigma_{YX}} - \boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1}\boldsymbol{\Sigma_{YX}} + \boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1}\boldsymbol{\Sigma_{YX}}$$
$$\text{Var}(\mathbf{U}) = \boldsymbol{\Sigma_{XX}} - \boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1}\boldsymbol{\Sigma_{YX}}$$
Notice that this exact result is defined in the problem as $\boldsymbol{\Sigma_{X|Y}}$.

3. **Determine the conditional distribution $f_{\mathbf{X|Y}}(\mathbf{x|y})$**
We can express $\mathbf{X}$ in terms of $\mathbf{U}$ and $\mathbf{Y}$:
$$\mathbf{X} = \mathbf{U} + \mathbf{A}\mathbf{Y}$$
We want to find the distribution of $\mathbf{X}$ given that $\mathbf{Y} = \mathbf{y}$. Because $\mathbf{U}$ and $\mathbf{Y}$ are independent, conditioning on $\mathbf{Y} = \mathbf{y}$ does not change the distribution of $\mathbf{U}$. It simply treats the $\mathbf{A}\mathbf{Y}$ term as a constant $\mathbf{A}\mathbf{y}$.
Therefore, conditionally on $\mathbf{Y} = \mathbf{y}$, $\mathbf{X}$ is an affine transformation of the Gaussian vector $\mathbf{U}$, meaning the conditional distribution is also Gaussian.
Conditional Mean ($\boldsymbol{\mu}^*$):
$$E[\mathbf{X} \mid \mathbf{Y} = \mathbf{y}] = E[\mathbf{U} + \mathbf{A}\mathbf{y} \mid \mathbf{Y} = \mathbf{y}]$$
$$E[\mathbf{X} \mid \mathbf{Y} = \mathbf{y}] = E[\mathbf{U}] + \mathbf{A}\mathbf{y}$$
Substitute $E[\mathbf{U}]$ and $\mathbf{A}$:
$$E[\mathbf{X} \mid \mathbf{Y} = \mathbf{y}] = (\boldsymbol{\mu_X} - \boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1}\boldsymbol{\mu_Y}) + \boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1}\mathbf{y}$$
$$\boldsymbol{\mu}^* = \boldsymbol{\mu_X} + \boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1}(\mathbf{y} - \boldsymbol{\mu_Y})$$
Conditional Variance:
$$\text{Var}(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) = \text{Var}(\mathbf{U} + \mathbf{A}\mathbf{y} \mid \mathbf{Y} = \mathbf{y})$$
Since $\mathbf{A}\mathbf{y}$ is a constant vector when $\mathbf{Y}$ is given, it has zero variance:
$$\text{Var}(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) = \text{Var}(\mathbf{U}) = \boldsymbol{\Sigma_{XX}} - \boldsymbol{\Sigma_{XY}}\boldsymbol{\Sigma_{YY}}^{-1}\boldsymbol{\Sigma_{YX}} = \boldsymbol{\Sigma_{X|Y}}$$
4. **Conclusion**
Since the conditional distribution is Gaussian with mean $\boldsymbol{\mu}^*$ and covariance matrix $\boldsymbol{\Sigma_{X|Y}}$, we can directly write its probability density function using the standard formula for a multivariate normal distribution defined on $\mathbb{R}^n$:
$$f_{\mathbf{X|Y}}(\mathbf{x|y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma_{X|Y}}|}} \exp\left( -\frac{1}{2}(\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma_{X|Y}}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$

1. **Define the Given Variables**
$X$ (Hidden Temperature): Represents the standardized temperature deviation. It follows a standard normal distribution.
$$X \sim \mathcal{N}(0, 1)$$
This means the mean $\mu_X = 0$ and the variance $\text{Var}(X) = 1$.
$Q$ (Mechanical Load): Represents the "noise" created by torque. It is independent of $X$ and follows a normal distribution.
$$Q \sim \mathcal{N}(0, 4)$$
This means the mean $\mu_Q = 0$ and the variance $\text{Var}(Q) = 4$.
$Y$ (The Measurement): The total observed spindle power, which is the sum of the temperature effect and the mechanical load.
$$Y = X + Q$$

2. **Determine the Distribution of $Y$**
Since $Y$ is a linear combination of two independent normal random variables ($X$ and $Q$), $Y$ is also normally distributed. We need to find its mean and variance.
Mean of $Y$:
Using the linearity of expectation:
$$E[Y] = E[X + Q] = E[X] + E[Q]$$
$$E[Y] = 0 + 0 = 0$$
So, $\mu_Y = 0$.
Variance of $Y$:
Since $X$ and $Q$ are independent, the variance of their sum is the sum of their variances:
$$\text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q)$$
$$\text{Var}(Y) = 1 + 4 = 5$$
Therefore, the distribution of the observed power $Y$ is:
$$Y \sim \mathcal{N}(0, 5)$$

3. **Find the Covariance and Joint Distribution of $X$ and $Y$**
To find the relationship between the hidden temperature $X$ and the measurement $Y$, we calculate their covariance.
$$\text{Cov}(X, Y) = \text{Cov}(X, X + Q)$$
Using the bilinearity of covariance:
$$\text{Cov}(X, Y) = \text{Cov}(X, X) + \text{Cov}(X, Q)$$
Since $\text{Cov}(X, X)$ is just $\text{Var}(X)$, and $X$ and $Q$ are independent (meaning $\text{Cov}(X, Q) = 0$):
$$\text{Cov}(X, Y) = \text{Var}(X) + 0$$
$$\text{Cov}(X, Y) = 1$$
Now we have all the components for the joint multivariate normal distribution of the vector $\mathbf{Z} = \begin{bmatrix} X \\ Y \end{bmatrix}$.
Mean Vector: $\boldsymbol{\mu} = \begin{bmatrix} E[X] \\ E[Y] \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$
Covariance Matrix: $\boldsymbol{\Sigma} = \begin{bmatrix} \text{Var}(X) & \text{Cov}(X, Y) \\ \text{Cov}(Y, X) & \text{Var}(Y) \end{bmatrix} = \begin{bmatrix} 1 & 1 \\ 1 & 5 \end{bmatrix}$
Therefore, the joint distribution is:
$$\begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathcal{N}\left( \begin{bmatrix} 0 \\ 0 \end{bmatrix}, \begin{bmatrix} 1 & 1 \\ 1 & 5 \end{bmatrix} \right)$$

4. **Calculate the Conditional Expectation $E[X \mid Y]$**
The primary goal in this scenario is to estimate the hidden internal temperature $X$ given a specific reading from the power sensor $Y = y$. The optimal estimate (which minimizes the mean squared error) is the conditional expectation $E[X \mid Y=y]$.
For jointly normal variables, the conditional expectation is given by the formula:
$$E[X \mid Y=y] = \mu_X + \frac{\text{Cov}(X, Y)}{\text{Var}(Y)} (y - \mu_Y)$$
Plugging in the values we calculated:
$\mu_X = 0$
$\text{Cov}(X, Y) = 1$
$\text{Var}(Y) = 5$
$\mu_Y = 0$

5. $$E[X \mid Y=y] = 0 + \frac{1}{5} (y - 0)$$
$$E[X \mid Y=y] = \frac{1}{5}y = 0.2y$$
